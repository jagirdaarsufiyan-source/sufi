###Common Method for Univariate Analysis - Numerical

```
def univariate_numerical(data):
  fig, axes = plt.subplots(3, 1, figsize=(10, 15))
  sns.histplot(data=data,ax=axes[0])
  sns.boxplot(data=data,orient='h',ax=axes[1])
  sns.kdeplot(data=data,ax=axes[2])
```

####Observations
####The count plot shows how different payment methods are used across various **product categories**. Some categories have higher counts, indicating they are more popular. The color hue division helps compare which **payment method** is most common within each category.


```
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Product_Category', hue='Payment_Method')
plt.title("Product Category by Country")
plt.xticks(rotation=45)
plt.show()
```

####Observations
#####The boxplot shows the price distribution across different product categories and payment methods. It highlights how prices vary within each category and which payment methods are linked to higher or lower prices. Outliers indicate unusually high or low prices in specific categories.


```
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Product_Category', hue='Payment_Method')
plt.title("Product Category by Country")
plt.xticks(rotation=45)
plt.show()
```

####Observations
#####The violin plot shows the distribution of payment methods within each product category. It combines the features of a boxplot and a density plot, displaying both the range and concentration of data. Wider sections indicate where most values are concentrated, while thinner parts show fewer observations.


```
sns.violinplot(data=data,x='Product_Category',hue='Payment_Method')
plt.show()
```

####Observations
The count plot shows how different quantities are distributed across various product categories. It helps identify which categories have higher sales volumes and which have lower ones. Taller bars indicate categories with more transactions or higher quantities sold.


```
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Product_Category', hue='Quantity')
plt.title("Product Category by Country")
plt.xticks(rotation=45)
plt.show()
```

####Observations
#####The pie chart shows the percentage share of each payment method in total sales. It clearly highlights which payment method is used most frequently and which are less common. Larger slices represent the more popular payment methods among customers.


```
plt.figure(figsize=(14,7))
data['Payment_Method'].value_counts().plot.pie(autopct='%1.1f%%',figsize=(8,8))
plt.title("Pie chart of Payment_Method Sales")
plt.show()
```
